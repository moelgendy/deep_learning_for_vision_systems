## Mini-project: Download MobileNet to classify dog breeds

### 1. import the required libraries

```python
import keras
from keras.applications import imagenet_utils
from keras.applications import MobileNet
from keras.applications.mobilenet import preprocess_input
from keras.preprocessing.image import ImageDataGenerator
from keras.preprocessing import image
import numpy as np
from IPython.display import Image
```

### 2. download mobilenet from Keras

```python
mobile = keras.applications.mobilenet.MobileNet()
```

```python
mobile.summary()
```

Just for the fun of it, let's print the names of all the layers in our network. This is another nice tool to review your network. More importantly, you can use it to pick specific layers. 

```python
for i,layer in enumerate(mobile.layers):
    if layer.trainable:
        print(i,layer.name)  
```

### 3. run inferences on a sample image

```python
def prepare_image(file):
    img_path = ''
    img = image.load_img(img_path + file, target_size=(224, 224))
    img_array = image.img_to_array(img)
    img_array_expanded_dims = np.expand_dims(img_array, axis=0)
    return keras.applications.mobilenet.preprocess_input(img_array_expanded_dims)
```

```python
Image(filename='images/german.jpeg')
```

```python
preprocessed_image = prepare_image('images/german.jpeg')
predictions = mobile.predict(preprocessed_image)
results = imagenet_utils.decode_predictions(predictions)
results
```
